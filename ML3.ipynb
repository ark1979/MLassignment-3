### Do we see correlations in feature space?
The below diagram shows the correlation between the features. When to features have high correlation it means that one affect the other positively, meaning if one goes up the second one follows. Correlations closer to 0 means that there are less of a link between them, if one increases the second one might increase but we don't know.

Below is a plot of the correlation between the data:

```python
fig, ax = plt.subplots(figsize=(15,11))
sns.heatmap(df.corr().round(2),linewidths=0.50,vmax=1.0, square=True, cmap="Blues", linecolor='black', annot=True, ax=ax)
```

We saw that some of the features had close to white colors in the heatmap showcasing a negative correlation. When features are negatively correlated it doesn't mean that there are no relation between them. However it means that when one feature increases the second feature decreases. So a negative correlation can imply that there is a relation between the features.

Below is a heatmap of the correlation between the features where all values are converted into absolute values:

```python
fig, ax = plt.subplots(figsize=(15,11))
sns.heatmap(df.corr().abs().round(2),linewidths=0.50,vmax=1.0, square=True, cmap="Blues", linecolor='black', annot=True, ax=ax)
```

With this heatmap we aim to show the features that might have a relation to one another so we do not distinquish between negative and positive correlation. If a feature have a small absolute correlation to other features it can either mean that the feature does not say a lot about the data or that the feature brings a new perspective on the table.
